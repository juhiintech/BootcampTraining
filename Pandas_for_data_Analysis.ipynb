```python
import pandas as pd
data = {
    "Name":['Arun','Varun','Karun','Narun','Marun'],
    "Age":[28,34,22,34,28],
    "Salary":[50000,60000,45000,52000,40000]
}
df = pd.DataFrame(data)
grouped = df.groupby("Age")["Salary"].sum()
print(grouped)
```

```python
# group multiple column
import pandas as pd
data = {
    "Name":['Arun','Varun','Karun','Narun','Marun'],
    "Age":[28,34,22,34,28],
    "Salary":[50000,60000,45000,52000,40000]
}
df = pd.DataFrame(data)
grouped = df.groupby(["Age","Name"])["Salary"].sum()
print(grouped)
```

```python
import pandas as pd
df_customers = pd.DataFrame({
    'CustomerID':[1,2,3],
    "Name":['Ram','Shyam','dhyamm']
})
df_orders = pd.DataFrame({
    'CustomerID':[1,2,4],
    "OrderAmount":[250,450,350]
})

df_merged = pd.merge(df_customers,df_orders,on="CustomerID",how="inner")
print('inner join')
print(df_merged)
```

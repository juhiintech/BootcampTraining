```python
import matplotlib.pyplot as plt
x = ['mon','tue','wed','thur','fri']
y = [10,15,7,20,12]
plt.plot(x,y)
plt.title('Bakery sales this week')
plt.xlabel('Day of the week')
plt.ylabel("Sales of the day")
plt.show()
```

```python
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000,1500,1200,1800]
plt.plot(months,sales,color='blue',linestyle='--',linewidth = 2,marker = 'o',label='2025 sales data')
plt.xlabel('Months')
plt.ylabel('Sales for month')
plt.title('Monthly Sales Data Report')
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='grey',linestyle = ':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4] ,{'M1','M2','M3','M4'}),
plt.show()
```

```python
# bar chart for comparison
import matplotlib.pyplot as plt

products = ['A','B','C','D']
sales = [1000,1500,800,1200]
plt.bar(products,sales,color='orange',label='sales 2026')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales Comparison')
plt.legend()
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors = ['gold','skyblue','lightgreen','coral'])
plt.title("Revenue contribution by region")
plt.show()
```

```python
# histogram
# continous data group or grouping continuing data.
import matplotlib.pyplot as plt
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]
plt.hist(scores,bins=5,color='purple',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of students')
plt.title('Score Distribution Of Students')
plt.show()
```

```python
# Scatterplots
# clustring
# relationship between two variables
import matplotlib.pyplot as plt
hours_studies = [1,2,3,4,5,6,7,8]
exams_scores = [50,55,60,65,70,75,80,85]
# scatter plot
plt.scatter(hours_studies,exams_scores,color='red',marker='o',label='student Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()


# find correlation between two variables
# used in machine learning alot in regression.
```

```python
import matplotlib.pyplot as plt
plt.scatter([1,2,3],[50,55,60],color='blue',label='Class A')
plt.scatter([1,2,3],[45,50,52],color='orange',label='Class B')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship between study time and exam score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# subplot
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]
plt.subplot(1,2,1)  #1row,2column,1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2) #1row,2column,2subplot
plt.bar(x,y)
plt.title('Bar Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,2,figsize = (10,5))
x = [1,2,3,4]
y = [10,20,15,25]
ax[0].plot(x,y)
ax[0].set_title('Line Plot')
ax[1].bar(x,y)
ax[1].set_title('Bar Chart')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,2,figsize = (10,5))
x = [1,2,3,4]
y = [10,20,15,25]
ax[0].plot(x,y,color='green')
ax[0].set_title('Line Plot')
ax[1].bar(x,y,color='red')
ax[1].set_title('Bar Chart')
plt.tight_layout()
plt.show()
```

```python
# save fig()
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,25]
# create plot
plt.plot(x,y,color='blue',marker ='o')
plt.title('Simple Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.savefig('Line_Plot.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
# Architecture of LSTM
# Layers and Shapes

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Hyperparameters
vocab_size = 1000        # Total unique words (tokens)
embedding_dim = 16       # Each word is represented by a 16-dimensional vector
sequence_length = 8      # Each input sentence has 8 words

# Model Architecture
lstm_model = Sequential(
    [
        # Layer 1: Embedding Layer
        Embedding(
            input_dim=vocab_size,
            output_dim=embedding_dim,
            input_length=sequence_length,
            name="embedding_layer"
        ),

        # Layer 2: LSTM Layer
        LSTM(
            units=32,
            name="lstm_layer"
        ),

        # Layer 3: Output Layer
        Dense(
            units=1,
            activation="sigmoid",
            name="output_layer"
        )
    ],
    name="Sequential_LSTM_Classifier"
)

# Build model (required before summary)
lstm_model.build(input_shape=(None, sequence_length))

# Model Summary
lstm_model.summary()
```

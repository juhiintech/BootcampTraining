```python
# broadcasting
prices = [100,200,300]
discount = 10
final_prices = []
for price in prices:
  final_price = price = (price*discount/100)
  final_prices.append(final_price)
print(final_price)
```

```python
# handling missing values
# np.isnan detect missing values
# np.nan_to_num()
# np.isnf()
# NaN not a number
```

```python
# np.isnan(array)
import numpy as np
arr = np.array([1,2,np.nan,4,np.nan,6])
print(np.isnan(arr))
print(np.nan == np.nan)
```

```python
# nan_to_num(array,nan=value) default - 0
import numpy as np
arr = np.array([1,2,np.nan,4,np.nan,6])
cleaned_arr = np.nan_to_num(arr,nan=100)
print(cleaned_arr)

```

```python
# np.isnf()
import numpy as np
arr = np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
```

```python
import numpy as np
arr = np.array([1,2,np.inf,4,-np.inf,6])
print(np.isinf(arr))
cleaned_arr = np.nan_to_num(arr,posinf=1000,neginf=1000)
print(cleaned_arr)
```

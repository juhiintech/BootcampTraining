```python
# TRANSFORMER
# Transformer makes self attention central.
'''
A Transformer is a deep learning architecture that uses self-attention to process all input tokens
simultaneously, enabling it to understand context efficiently. It is faster than
RNNs and LSTMs and forms the foundation of modern large language models.
'''
```

```python
# Encoder and Decoder
'''
An Encoder is a neural network that takes the input sequence,
understands its meaning and context, and converts it into a meaningful
numerical representation (called embeddings or hidden representations).


A Decoder is a neural network that takes the encoder's output and generates
the desired output one token at a time.



An encoder is the part of a Transformer that reads and understands the
input sequence by converting it into meaningful representations.
A decoder uses those representations (or previous context in decoder-only models)
to generate the output sequence one token at a time.
'''
```

```python
# Embedding
'''
An embedding is a numerical vector that represents the meaning of a word,
sentence, image, or other data in a way that a machine learning model can understand.
'''
```

```python
# token embedding
# A token embedding is the dense numerical vector representation of a token.
# It converts text into numbers that capture semantic meaning, allowing Transformer
# models to process and understand language.
```

```python
'''
Text
   ↓
Tokenization
   ↓
Tokens
   ↓
Token Embeddings
   ↓
Transformer
'''
```

```python
# unigram
'''
A unigram auto-suggest model predicts words based only on their overall
frequency in the training data. It does not consider the previous words or
sentence context, making it the simplest form of language model.
'''
```

```python
# Positional Encoding
'''
Positional Encoding is a technique used in Transformers to add position
information to token embeddings, allowing the model to understand the order
of words in a sequence because Transformers process all tokens in parallel.
'''
```

```python
# Repetitive Transformer Blocks
'''
A repetitive Transformer block is the repeated stacking of identical Transformer
layers. Each block contains self-attention and a feed-forward network, and each
successive block refines the representation learned by the previous one,
enabling the model to understand increasingly complex patterns in the data.
'''
```

```python
'''
Number of Transformer Blocks

Different models use different numbers of blocks.

Examples:

BERT Base → 12 Transformer blocks
BERT Large → 24 Transformer blocks
GPT-2 → up to 48 Transformer blocks (depending on the model size)
Modern large language models may use dozens to hundreds of Transformer blocks.

Generally, more blocks allow the model to learn more complex patterns, but
they also require more computation and memory.
'''
```

```python
# Multi-head self attention
'''
Multi-Head Self-Attention is a mechanism in Transformers where multiple
attention heads simultaneously learn different relationships between tokens,
helping the model understand context more effectively.
'''
```

```python
# Temperature and How Tired Are You
# Simple Perceptron Example

# -----------------------------
# Inputs
# Is it a nice temperature? (Yes/No)
# Are you tired? (Yes/No)
# -----------------------------

temperature = input("Is it a nice temperature? (Yes/No): ").strip().lower()
tired = input("Are you tired? (Yes/No): ").strip().lower()

# Convert Yes/No into 1/0
temperature_input = 1 if temperature == "yes" else 0
tired_input = 1 if tired == "yes" else 0

# -----------------------------
# Weights (Importance)
# -----------------------------
temperature_weight = 5
tired_weight = -2

# -----------------------------
# Bias
# Likes running = +3
# Dislikes running = -4
# -----------------------------
bias = 3   # Change to -4 if the person dislikes running

# -----------------------------
# Weighted Sum
# -----------------------------
temperature_score = temperature_input * temperature_weight
tired_score = tired_input * tired_weight

output = temperature_score + tired_score + bias

# -----------------------------
# Results
# -----------------------------
print("\n----- Calculation -----")
print(f"Temperature: {temperature_input} × {temperature_weight} = {temperature_score}")
print(f"Tiredness : {tired_input} × {tired_weight} = {tired_score}")
print(f"Bias = {bias}")
print(f"Total Score = {output}")

# -----------------------------
# Decision
# -----------------------------
if output > 0:
    print("Decision: ✅ Go for a Run!")
else:
    print("Decision: ❌ Don't Go for a Run.")
```

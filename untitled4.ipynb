```python
#working with real world data for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df = pd.read_csv(url)

print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df = df.rename(columns={
    "lifeExp": "life_expectancy",
    "pop": "population",
    "gdpPercap": "gdp_per_capita"
})

print(df.head())
print(df.columns)
```

```python
print("Missing values in each column: ")
print(df.isnull().sum())


print("Number of duplicate rows: ", df.duplicated().sum())
```

```python
dirty_demo = df.head(12).copy()
#deliberately introduce missing values for teaching practice

dirty_demo.loc[2,"life_expectancy"]=np.nan
dirty_demo.loc[5,"gdp_per_capita"]=np.nan

print(dirty_demo)
print(dirty_demo.isna().sum())

#fill mising numerical values using median

clean_demo = dirty_demo.copy()
clean_demo["life_expectancy"]=clean_demo["life_expectancy"].fillna(clean_demo["life_expectancy"].median())
clean_demo["gdp_per_capita"]=clean_demo["gdp_per_capita"].fillna(clean_demo["gdp_per_capita"].median())


print(clean_demo.isna().sum())
print(clean_demo)
```

```python
#using pandas-- i want all of you to create a basic data analysis-- #you can take use of lm

#year range
#number of countries
#number of continents

print("Year range:", df["year"].min(), "to", df["year"].max())
print("Number of countries:", df["country"].nunique())
print("Number of continents:", df["continent"].nunique())

print("Average life expectancy:", df["life_expectancy"].mean())
print("Minimum life expectancy:", df["life_expectancy"].min())
print("Maximum life expectancy:", df["life_expectancy"].max())

print(df.describe())
```

```python
#feature extraction
#groups and counts-(gdp population life expectancy)
print(df["continent"].value_counts())

continent_summary = df.groupby("continent")[[
      "life_expectancy",
    "gdp_per_capita",
    "population"
]].mean()

print(continent_summary)
```

```python
latest_year = df["year"].max()
latest_df = df[df["year"] == latest_year].copy()

print("Latest year:", latest_year)
print(latest_df.head())
print(latest_df.shape)
```

```python
#using the lib matplotlib #I need a bar chart that shows me avg life expectancy by continent

continent_life = latest_df.groupby("continent")["lifeExp"].mean().sort_values()

plt.figure(figsize=(6, 5))
plt.bar(continent_life.index, continent_life.values)

plt.title(f"Average Life Expectancy by Continent in {latest_year}")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#line chart life expectancy over time for 3 countries India, China, United States

selected_countries = ["India", "China", "United States"]
trend_df = df[df["country"].isin(selected_countries)]
plt.figure(figsize=(9, 5))

for country in selected_countries:
    country_df = trend_df[trend_df["country"] == country]

    plt.plot(
        country_df["year"],
        country_df["lifeExp"],
        marker="o",
        label=country
    )

plt.title("Life Expectancy Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")

plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
#scatter plot GDP per capita vs life expectancy
# 2007 ka data filter
latest_df = df[df["year"] == 2007]

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(
    latest_df["gdp_per_capita"],
    latest_df["life_expectancy"],
    alpha=0.7
)
# Log scale on x-axis
plt.xscale("log")
# Labels and title
plt.title("GDP per Capita vs Life Expectancy in 2007")
plt.xlabel("GDP per Capita - log scale")
plt.ylabel("Life Expectancy")

plt.grid(True)

plt.show()
```

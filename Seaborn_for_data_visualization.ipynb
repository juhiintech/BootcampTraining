```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
y_1 = sns.load_dataset('penguins').head(20)
# x_1 = pd.DataFrame({"var":var,"var_1":var_1})
# sns.lineplot(x="var",y="var_1",data=x_1)
y_1
```

```python
# barplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",palette="icefire",
            saturation=6,errcolor="g",capsize=0.5,dodge=True,alpha=0.2)
plt.show()
```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
            order = order_1,hue_order=["Female","Male"],ci=100,n_boot=2)
plt.show()


```

```python
order_1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=var,orient="h")

plt.show()
```

```python
# Scatter plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("penguins").head(20)
var
```

```python
m = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,
                hue="sex",style="sex",size="sex",sizes=(100,40),palette="gist_ncar",
                alpha=1,markers=m)
plt.show()
```

```python
# barplot in seaborn
# barplot plots a value or metric for each category(by default,barplot plots the mean of a variable by category)
```

```python
# count plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=var,hue="smoker",palette="bwr",color="r")
```

```python
# Violin plot in seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time",
               linewidth=3,palette="Dark2",saturation=0.2,color="y")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,hue="sex",split=True,order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="count")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var,hue="sex",split=True,scale="count")
plt.show()
```

```python
# pair plot in seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn",kind="reg")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"],palette="BuGn",kind="kde")
plt.show()
```

```python
# strip plot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=var,hue="sex",
              palette="winter_r",linewidth=1,edgecolor="r",
              jitter=3,size=5)
plt.show()
```

```python
# Box plot in seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=var,hue="sex",color="g",palette="plasma")
plt.show()
```

```python
# factor plot in seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
var = sns.load_dataset("tips")
var
```

```python
sns.catplot(x="size",y="tip",data=var) #factorplot...>catplot
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex") #factorplot...>catplot
plt.show()
```
